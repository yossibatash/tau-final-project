```python
import time
from datetime import datetime, timedelta

base_path = "/data/alpha_vantage/news_data"

for c in companies:
    
    start_date = '2022/02/27'
    ticker = c
    company_name = companies[c]

    # checking if the directory demo_folder 
    # exist or not.   
    output_path = f"{base_path}/{ticker}"
    if not os.path.exists(output_path):      
        # if the demo_folder directory is not present 
        # then create it.
        os.makedirs(output_path)    
    
    
#     print(f"Company name is {company_name}, and Ticker is -{ticker}")
    print(f"-------------Start Extracting News Data for - {company_name}/{ticker}")
    for week in range(0,52):   
        date = datetime.strptime(start_date, "%Y/%m/%d")
        modified_date = date + timedelta(days=7)
        end_date = datetime.strftime(modified_date, "%Y/%m/%d")
        sd = start_date.replace("/","")
        ed = end_date.replace("/","")      
        # Preper API request
        CSV_URL = f"https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={ticker}&time_from={sd}T0005&time_to={ed}T0005&limit=200&apikey={apikey}"
        print(CSV_URL)
        with requests.Session() as s:
            download = s.get(CSV_URL)
            ouput_file = f"{output_path}/news_data_{sd}_{ed}.csv"
            try:
                ds = json.loads(download.content)["feed"]
                df = pd.DataFrame(ds)
                df.to_csv(path_or_buf=ouput_file,
                          quotechar='"',
                          index=False,
                          quoting=csv.QUOTE_ALL)       
            except Exception as e:
                print(f"Error: {e}")
        print(f"Finished extracting {ticker} news data between {start_date} and {end_date}, now sleeping for 5 minutes")
        for i in range(300, 0, -1):
            print(f"Weaking up in - {i} sec", end = '\r')
            time.sleep(1)                  
        start_date = end_date            

```
