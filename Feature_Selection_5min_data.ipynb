```python
# Correct the column assignment, rename the columns so that they are named PC1, PC2, ...
pca_features = pd.DataFrame(pca_features, index = scaled_features.index, columns=[f"PC{i+1}" for i in range(pca_features.shape[1])])
```

```python
# --- Analyze Loadings  ---
loadings = pd.DataFrame(pca.components_.T, index=all_features, columns = [f"PC{i+1}" for i in range(pca_features.shape[1])])
```

```python
loadings
```

```python
top_n = 20
selected_features = set()
for i in range(pca_features.shape[1]): # for each principal component
    component_loadings = loadings[f"PC{i+1}"].abs()
    top_features = component_loadings.nlargest(top_n).index
    selected_features.update(top_features) # This line is important, it adds multiple feature to the set
selected_features = list(selected_features)

# I played with top_n values until I managed to reduce the selected_features to ~ 50%  
print(f"Count of selected_features - {len(selected_features)}")
print(selected_features)
```
